```python
import pandas as pd
```

```python
home_basic = pd.read_csv(r'C:\Users\dvale\TFM\Data\home_basic.csv')
```

```python
home_basic.head()
```

```python
home_basic.rename(columns={'DB010': 'Year of survey'}, inplace=True)

home_basic.head()
```

```python
home_basic.rename(columns={'DB020': 'Country'}, inplace=True)

home_basic.head()
```

```python
home_basic.rename(columns={'DB030': 'Household identification number'}, inplace=True)

home_basic.head()
```

```python
home_basic.rename(columns={'DB040': 'Region'}, inplace=True)

mapped_region = {'ES11': 'Galicia', 
               'ES12': 'Asturias', 
               'ES13': 'Cantabria', 
               'ES21': 'Basque Country', 
               'ES22': 'Navarre',
               'ES23': 'La Rioja',
               'ES24': 'Aragon',
               'ES30': 'Comunity of Madrid',
               'ES41': 'Castile and Leon',
               'ES42': 'Castileâ€“La Mancha',
               'ES43': 'Extremadura',
               'ES51': 'Catalonia',
               'ES52': 'Valencian Community',
               'ES53': 'Balearic Islands',
               'ES61': 'Andalusia',
               'ES62': 'Region of Murcia',
               'ES63': 'Ceuta',
               'ES64': 'Melilla',
               'ES70': 'Canary Islands',
               'ESZZ': 'Extra-regional'
              }

home_basic['Region'] = home_basic['Region'].map(mapped_region)

home_basic.head()
```

```python
home_basic.rename(columns={'DB040_F': 'Region (flag)'}, inplace=True)

mapped_region_flag = {-1: 'Not reported', 1: 'Reported'}

home_basic['Region (flag)'] = home_basic['Region (flag)'].map(mapped_region_flag)

home_basic.head()
```

```python
home_basic.rename(columns={'DB060': 'Primary sample unit'}, inplace=True)

home_basic.head()
```

```python
home_basic.rename(columns={'DB060_F': 'Primary sample unit (flag)'}, inplace=True)

mapped_sample_flag = {-1: 'Not reported', 2: 'Reported'}

home_basic['Primary sample unit (flag)'] = home_basic['Primary sample unit (flag)'].map(mapped_sample_flag)

home_basic.head()
```

```python
home_basic.rename(columns={'DB090': 'Household cross factor value'}, inplace=True)

home_basic.head()
```

```python
home_basic.rename(columns={'DB090_F': 'Household cross factor value (flag)'}, inplace=True)

home_basic['Household cross factor value (flag)'] = home_basic['Household cross factor value (flag)'].map({1: 'Reported'})

home_basic.head()
```

```python
home_basic.rename(columns={'DB100': 'Population density'}, inplace=True)

mapped_population = {1: 'High density', 2: 'Medium density', 3: 'Low density'}

home_basic['Population density'] = home_basic['Population density'].map(mapped_population)

home_basic.head()
```

```python
home_basic.rename(columns={'DB100_F': 'Population density (flag)'}, inplace=True)

home_basic['Population density (flag)'] = home_basic['Population density (flag)'].map({1: 'Reported'})

home_basic.head()
```
