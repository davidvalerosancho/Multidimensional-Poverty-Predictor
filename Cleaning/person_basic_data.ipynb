```python
person_basic.rename(columns={'RB210': 'Working status'}, inplace=True)

working_dict = {'1': 'Employed', '2': 'Unemployed', '3': 'Retired', '4': 'Other type of economic inactivity'}

working_column = []

for element in person_basic['Working status']:
    if element not in working_dict.keys():
        working_column.append('Not reported')
    else:
        for key in working_dict.keys():
            if element == key:
                working_column.append(working_dict[key])

person_basic['Working status'] = working_column

person_basic[person_basic.columns[5:]].head()
```
