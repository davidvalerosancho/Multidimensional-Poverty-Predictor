```python
import pandas as pd
```

```python
left = pd.read_csv(r'C:\Users\dvale\TFM\Data\sociodemo.csv')
left = left.drop(['Unnamed: 0', 'material_deprivation'], axis=1)

left.head()
```

```python
rigth = pd.read_csv(r'C:\Users\dvale\TFM\Data\occupation.csv')
rigth = rigth.drop(['Unnamed: 0', 'weight', 'material_deprivation'], axis=1)

rigth.head()
```

```python
all_variables = left.merge(rigth, how='left', on='id')
all_variables.head()
```

```python
rigth = pd.read_csv(r'C:\Users\dvale\TFM\Data\other.csv')
rigth = rigth.drop(['Unnamed: 0', 'weight', 'material_deprivation'], axis=1)

rigth.head()
```
